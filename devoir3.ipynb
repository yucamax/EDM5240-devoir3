```python
ati.pages.count()
#ceci représente le nombre total de pages demandées par les gens ayant fait une demande d'information à l'entièreté des entités.
```

```python
ati.owner_org_title.value_counts()
#cette formule montre en ordre décroissant le nombre de demandes d'accès à l'information qui ont été faites à chacune des entités.
```

```python
fin = ati.owner_org_title == "Department of Finance Canada | Ministère des Finances Canada"
ati[fin].disposition.value_counts()
#ceci nous permet de voir la quantité de documents qui ont été donnés partiellement, aucunement ou en totalité par 
#le ministère des finances aux auteurs des demandes d'accès à l'information.
```

```python
ati.disposition.value_counts()
#Avec cette formule, nous pouvons combien de documents ont été donnés partiellement, en entièreté ou pas du tout à
#l'auteur de la demande d'accès à l'information. Cela nous permet aussi de voir que les formulations ne sont pas toutes
#sembables.
```

```python
#Question de départ : Quelle est l'entité qui donne le plus de pages d'informations?
#Ici nous allons voir, en ordre décroissant, quelle entité a livré le plus de pages d'informations aux auteurs des demandes.
ati.groupby("owner_org_title").pages.sum().sort_values(ascending=False)
```

```python
ati.owner_org_title.value_counts()/len(ati.owner_org_title)*100
#cette variable permet de voir en pourcentage le nombre de demandes d'accès faites à chaque entité
```
